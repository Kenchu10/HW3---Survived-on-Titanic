```python
# Get the corelation of different fields

df.corr()
```

```python
# Do the regression With intercept

import statsmodels.formula.api as sm

result=sm.ols(formula='survived ~ sex + fare + pclass', data=df).fit()
result.summary()
```

```python
# Do the regression Without intercept


result=sm.ols(formula='survived ~ sex +  fare + pclass -1 ', data=df).fit()
result.summary()
```

```python
# Do the regression Without intercept on gender alone


result=sm.ols(formula='survived ~ sex -1', data=df).fit()
result.summary()
```

```python

# It seemed that regression without intercept has better fit
```

```python
# make new columns for Pivot tables

df['Passengers'] = "Passenger"

df['Gender'] = np.where(df['sex']==0, 'male','female')

df['Survived'] = np.where(df['survived']==1, 'Yes','No')


df.head()
```

```python
# Do the Pivot table on passengers

pd.pivot_table(df, index=['Gender',], aggfunc='count', values=["Passengers"], margins=True)
```

```python
# Do the Pivot table on passengers to show survival

pd.pivot_table(df, index=['Gender','Survived'], aggfunc='count', values=["Passengers"], margins=True)
```

```python
# Do the Pivot table on passengers to show class cabins and survival

pd.pivot_table(df, index=['pclass','Survived'], aggfunc='count', values=["Passengers"], margins=True)
```

```python
# Do the Pivot table on passengers to show gender, class cabins and survival

pd.pivot_table(df, index=['Gender','pclass','Survived'], aggfunc='count', values=["Passengers"], margins=True)
```

```python
# Do the Pivot table on gender, pclass of all survived

pd.pivot_table(df[survived], index=['Gender','pclass','Survived'], aggfunc='count', values=["Passengers"], margins=True)
```

```python
# Do the Pivot table on gender, pclass of all died

pd.pivot_table(df[died], index=['Gender','pclass','Survived'], aggfunc='count', values=["Passengers"], margins=True)
```
